```python
def get_bigram(word_list):
    tuple_list = [] 
    # to avoid "ValueError: all the input array dimensions
    # except for the concatenation axis must match exactly"
    
    if len(word_list) <= 1:
        return tuple_list
    
    for i in range(len(word_list) - 1):
        if len(word_list[i]) <= 3:
            continue
        for j in range(i + 1, len(word_list)):
            if len(word_list[j]) <= 3:
                continue
            set_temp = set([word_list[i], word_list[j]])
            if len(set_temp) > 1:
                tuple_list.append(str(tuple(set_temp)))

    return tuple_list
```

```python
from ast import literal_eval
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt
from itertools import groupby


def save_wordcloud(df, name = 'test.png'):
    if type(df['OFFENSE_DESCRIPTION_WORDS'][0]) == str:
        word_list = np.hstack(list(df['OFFENSE_DESCRIPTION_WORDS'].apply(literal_eval)))
    else:
        word_list = np.hstack(list(df['OFFENSE_DESCRIPTION_WORDS']))
    word_list_clean = [w for w in word_list if len(w) > 3] #consider words less than 3 characters are not meaningful
    freq_dict = dict((k, len(list(xs))) for k, xs in groupby(sorted(word_list_clean)))
    
    wordcloud = WordCloud(width=1280, height=720, background_color='white').fit_words(freq_dict)
    wordcloud.to_file(name)
```

```python
save_wordcloud(df, 'test3.png')
```
