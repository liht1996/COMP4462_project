```python
df_w = pd.read_csv("1685018.csv", parse_dates=['DATE'])
```

```python
df_sod = df_w[df_w['REPORT_TYPE']=='SOD  ']
```

```python
df_sod.filter(regex ='Daily|DATE').to_csv('weather.csv')
```

```python
df_sod.filter(regex ='Daily|DATE').columns
```

```python
df_sod_f = df_sod.filter(regex ='Daily|DATE')
```

```python
from datetime import datetime

df_sod_f['Date_new'] = df_sod_f['DATE'].apply(lambda x: datetime(x.year, x.month, x.day))
```

```python
df_sod_f.head(10)
```

```python
df_sod_f.groupby('DailyWeather').count()
```

```python
df_sod_f['DailyWeather'] = df_sod_f['DailyWeather'].fillna('Sunny')
```

```python
def weather(x):
    if 'RA' in x:
        return "Rainy"
    elif 'SN' in x:
        return "Snowy"
    elif x == "Sunny":
        return "Sunny"
    else:
        return "Other"
```

```python
df_sod_f['Max_Temp'] = df_sod_f['DailyMaximumDryBulbTemperature']
df_sod_f['Min_Temp'] = df_sod_f['DailyMinimumDryBulbTemperature']
df_sod_f['Ave_Temp'] = df_sod_f['DailyAverageDryBulbTemperature']


df_sod_f['Weather'] = df_sod_f['DailyWeather'].apply(weather)
```

```python
df_sod_f.groupby('Weather').count()
```

```python
df_sod_f_s = df_sod_f[['Date_new', 'Max_Temp', 'Min_Temp', 'Ave_Temp', 'Weather']]
```

```python
df['Date_new'] = df['OCCURRED_ON_DATE'].apply(lambda x: datetime(x.year, x.month, x.day))
```

```python
df = df.merge(df_sod_f_s, on = 'Date_new')
```

```python
df.to_csv('crimes_group_word_weather.csv')
```

```python
df.to_json('data.json')
```
